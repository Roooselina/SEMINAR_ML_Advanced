```python
import numpy as np
```

```python
# 입력 데이터

X = np.random.rand(1000000, 1)
y = 2 * X + 1 + np.random.randn(1000000, 1) * 0.1
```

```python
import matplotlib.pyplot as plt

# 데이터 시각화
plt.scatter(X, y, label='Data points')
plt.xlabel('X')
plt.ylabel('y')
plt.title('maked data')
plt.legend()
plt.show()
```

```python
import math

# 데이터를 미니 배치에 담기

batch_size = 512
num_batches = math.ceil(len(X) / batch_size)

def data_loader(data, batch_size, num_batches):
    loaded_data = []
    for i in range(num_batches):
        start_index = i * batch_size
        end_index = min((i + 1) * batch_size, len(data))
        mini_batch = data[start_index:end_index]
        loaded_data.append(list(mini_batch))  # 리스트로 변환하여 추가
    return loaded_data


# 데이터를 미니 배치로 나누기
X_batches = data_loader(X, batch_size, num_batches)
y_batches = data_loader(y, batch_size, num_batches)

print("X; length of mini batch:", X_batches[0][0])  # 전체 미니배치의 개수 출력
print("X; batch size:", len(X_batches[10]))  # 11번째 미니배치의 길이 출력
print("y; length of mini batch:", (y_batches[0]))  # 전체 미니배치의 개수 출력
print("y; batch size:", len(y_batches[10]))  # 11번째 미니배치의 길이 출력


print("length of X:",len(X))

## x는 임의의 100개 데이터가 저장되어 있다
```

```python
# 손실 함수 정의 (평균 제곱 오차)
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
# 선형 함수 정의: y = wx + b
def linear_function(X, w, b):
    return np.dot(X, w) + b
```

```python
import time

# 경사 하강법 함수 정의
def gradient_descent(X_batches, y_batches, learning_rate=0.001, epochs=100):
    start = time.time()
    costs = []
    
    num_batches = len(X_batches)
    num_samples = sum(len(batch) for batch in X_batches)
    num_features = len(X_batches[0][0])  # 첫 번째 미니 배치의 첫 번째 샘플의 특성 수를 기준으로 함
    
    w = np.random.randn(num_features, 1)  # 가중치 초기화
    b = np.random.randn(1)  # 편향 초기화
    
    # 경사 하강법으로 파라미터 업데이트
    for epoch in range(epochs):
        epoch_cost = 0
        for X_batch, y_batch in zip(X_batches, y_batches):
            X_batch = np.array(X_batch)  # 리스트를 넘파이 배열로 변환
            y_batch = np.array(y_batch)  # 리스트를 넘파이 배열로 변환
            
            # 예측값 계산
            predictions = np.array([linear_function(x, w, b) for x in X_batch])  # 각 샘플에 대해 예측값 계산
            
            # 오차 계산
            error = predictions - y_batch
            mse = np.mean(error ** 2)  # MSE 계산
            epoch_cost += mse
            
            # 경사도 계산
            dw = (1 / len(X_batch)) * np.dot(X_batch.T, error)
            db = (1 / len(X_batch)) * np.sum(error)
                
            # 경사 하강법을 사용하여 파라미터 업데이트
            w -= learning_rate * dw
            b -= learning_rate * db
            
            # 손실 함수 값 저장
            costs.append(epoch_cost / num_samples)
    
        # 손실 함수 값 출력
        if epoch % 10 == 0:
            print(f"Epoch {epoch+1}: Average MSE: {epoch_cost / num_samples}")

        # 미니 배치의 첫 번째 요소만 전달하여 경사도 검사
        check_gradients(X_batch[0], y_batch[0], w, b, epoch)
    
    end = time.time()
    elapsed_time = end - start  # 경과 시간 계산
    return w, b, elapsed_time, costs
```

```python
def check_gradients(X, y, w, b, epoch):
    epsilon = 1e-7

    # 가중치에 대한 기울기 계산
    grad_w = np.zeros_like(w)
    for i in range(w.shape[0]):
        for j in range(w.shape[1]):
            w_plus = w.copy()
            w_minus = w.copy()
            w_plus[i, j] += epsilon
            w_minus[i, j] -= epsilon
            grad_w[i, j] = (mse_loss(y, linear_function(X, w_plus, b)) - mse_loss(y, linear_function(X, w_minus, b))) / (2 * epsilon)

    # 편향에 대한 기울기 계산
    grad_b = np.zeros_like(b)
    for i in range(b.shape[0]):
        b_plus = b.copy()
        b_minus = b.copy()
        b_plus[i] += epsilon
        b_minus[i] -= epsilon
        grad_b[i] = (mse_loss(y, linear_function(X, w, b_plus)) - mse_loss(y, linear_function(X, w, b_minus))) / (2 * epsilon)

    if i == w.shape[0]-1:
        if epoch % 10 == 0:
            if grad_w<1e-7:
                print("제대로 동작중 good", grad_w)
            else:
                print("오류 있는듯 ㅠㅜ", grad_w)
```

```python
# 경사 하강법을 사용하여 모델 학습
w, b, time, cost = gradient_descent(X_batches, y_batches)

print("Final parameters:")
print("w:", w)
print("b:", b)
print("time:", time)


def draw_linear(X, w, b):
    return w * X + b

# 일직선 그래프 그리기
plt.scatter(X, y, label='Data points')
plt.plot(X, draw_linear(X, w, b), color='red', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Data and Linear Regression Graph')
plt.legend()
plt.show()


# 손실 함수 값 그래프 그리기
plt.plot(cost)
plt.xlabel('Epochs')
plt.ylabel('Cost')
plt.title('Cost per Epoch')
plt.show()
```

```python
#미분 정확도 비교

# 미분할 함수 정의
def f(x):
    return x ** 2

# 수학적 미분 계산
def df_exact(x):
    return 2 * x

# 수치 미분 계산 1
def df_numerical(x, epsilon=1e-6):
    return (f(x + epsilon) - f(x - epsilon)) / (2 * epsilon) - (2 * x)

# 수치 미분 계산 2
def df_numerical_2(x, epsilon=1e-6):
    return (f(x + epsilon) - f(x)) / (epsilon) - (2 * x)

# x 범위 설정
X = np.linspace(-10, 10, 1000)

# 수치 미분 값 계산
numerical_gradients = df_numerical(X)
numerical_gradients2 = df_numerical_2(X)

# 결과 그래프 그리기
plt.plot(X, numerical_gradients, label='Numerical gradients', linestyle='--', color='red')
plt.plot(X, numerical_gradients2, label='Numerical gradients2', linestyle='--', color='blue')
plt.title('Comparison of numerical gradients')
plt.xlabel('X')
plt.ylabel('Gradient')
plt.legend()
plt.grid(True)
plt.show()
```
